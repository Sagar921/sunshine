```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a= np.array([1,2,3,4,5,6,7])
b= np.array([1,2,3,4,5,6,7])
```

```python
plt.title("Linear Line")
plt.xlabel("a")
plt.ylabel("b")
plt.plot(a,b)
plt.show()
```

```python
c= np.array([1,2,3,4,5,6,7])
d= np.array([11,21,31,41,51,61,71])
plt.plot(c,d)
plt.xlabel("c")
plt.ylabel("d")
plt.title("line Graph")
plt.show()
```

```python
plt.plot(c,d,'--k')
plt.show()
```

```python
x= np.array([5,10,15])
y= np.array([20,24,28])
plt.bar(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Bar Graph")
plt.show()
```

```python
plt.bar(x,y,color='g', align = 'center')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Bar Graph")

```

```python
a= np.array([22,10,12,422,62,72,82,92])
plt.hist(a,bins =[0,20,40,60,80,100])
plt.title("Histogram")
plt.show()
```
